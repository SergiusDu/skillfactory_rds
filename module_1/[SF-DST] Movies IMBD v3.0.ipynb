# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
d31 = data.copy()
d31['Title_Len'] = d31['original_title'].apply(lambda row: len(row))
d31['production_companies'] = d31['production_companies'].apply(lambda row: row.split('|'))
c_len_by_company = Counter()
c_count_of_films_by_company = Counter()
def count_names_len_and_films_count_by_company(row):
    for company in row['production_companies']:
        c_len_by_company[company] += len(row['original_title'])
    for company in row['production_companies']:
        c_count_of_films_by_company[company]+=1

d31.apply(lambda row:count_names_len_and_films_count_by_company(row), axis=1)
c_len_by_company.most_common()
c_count_of_films_by_company.most_common()

d31_first_part = pd.DataFrame(c_len_by_company.most_common(), columns=['Company_name','lenght'])
d31_first_part = d31_first_part.set_index('Company_name')
d31_second_part = pd.DataFrame(c_count_of_films_by_company.most_common(), columns=['Company_name','Count']).set_index('Company_name')

d31_final = d31_first_part.join(d31_second_part)
d31_final['mean_name_langht'] = d31_final['lenght'] / d31_final['Count']
d31_final.sort_values('mean_name_langht', ascending=False)
```

```python
answer_ls.append(5)
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
d36 = data.copy()
d36['profit'] = d36.apply(lambda row: row['revenue'] - row['budget'], axis=1)
d36['isprofitable'] = d36['profit'].apply(lambda row: 1 if row > 0 else 0)
d36['director'] = d36['director'].apply(lambda row: row.split('|'))

films_with_profit_by_director = Counter()
total_films_by_director = Counter()
def count_result(row):
    for person in row['director']:
        total_films_by_director[person]+=1
        if row['isprofitable'] == 1:
            films_with_profit_by_director[person]+=1
        else:
            pass
d36.apply(lambda row: count_result(row), axis=1)

df = pd.DataFrame(total_films_by_director.most_common(), columns=['director', 'total_films']).set_index('director')
df_profitable = pd.DataFrame(films_with_profit_by_director.most_common(), columns=['director', 'films_with_profit']).set_index('director')
df = df.join(df_profitable, how='left').fillna(0)
def calculate_chances(row):
    if row['films_with_profit']==0:
        pass
    else:
        return row['films_with_profit']/row['total_films']


df['chance_be_profitable'] = df.apply(lambda row: calculate_chances(row), axis=1).fillna(0)
df.sort_values('chance_be_profitable', ascending=False)

list_of_directors = [
    'Quentin Tarantino',
    'Steven Soderbergh',
    'Robert Rodriguez',
    'Christopher Nolan',
    'Clint Eastwood'
]

df[df.index.isin(list_of_directors)].sort_values('chance_be_profitable', ascending=False)
```
